```python
import pandas as pd
url = 'https://raw.githubusercontent.com/james-d-kaye-92/data-science-obesity/refs/heads/main/ObesityDataSet_raw_and_data_sinthetic.csv'
df1 = pd.read_csv(url)
```

```python
import numpy as np
from sklearn import preprocessing
```

```python
df1.head()
```

```python
df1 = df1.rename(columns={'family_history_with_overweight':'family_history_with_obesity', 'FAVC':'high_calory_food', 'FCVC':'vegetables', 'NCP':'number_of_meals', 'CAEC':'eats_food_between_meals', 'SMOKE':'smoke', 'CH2O': 'water', 'SCC':'monitor_calories', 'FAF':'exercise','TUE':'device_use','CALC':'alcohol','MTRANS':'transport','NObeyesdad':'obesity_level','Gender':'gender','Age':'age', 'Height':'height','Weight':'weight' })
```

```python
df1.head()
```

```python
age_values = df1['age'].values.tolist()
print(age_values)
unique_age_values = set(age_values)
print(unique_age_values)
```

```python
age_column = df1['age']

def rounder_age(x):
  return np.round(x).astype(int)
new_age_column = age_column.apply(rounder_age)
unique_new_age_column = set(new_age_column)

print(new_age_column.values.tolist())
print(unique_new_age_column)
print(len(unique_new_age_column))
```

```python
df2 = df1
df2['age'] = new_age_column
df2.head()
```

```python
height_values = df1['height'].values.tolist()
print(height_values)
unique_height_values = set(height_values)
print(unique_height_values)
```

```python
height_column = df1['height']

def rounder_height(x):
  return np.round(x,2)
new_height_column = height_column.apply(rounder_height)
unique_new_height_column = set(new_height_column)

print(new_height_column.values.tolist())
print(unique_new_height_column)
print(len(unique_new_height_column))
```

```python
df2['height'] = new_height_column
df2.head()
```

```python
weight_column = df1['weight']

def rounder_weight(x):
  return np.round(x,1)
new_weight_column = weight_column.apply(rounder_weight)
unique_new_weight_column = set(new_weight_column)

print(new_weight_column.values.tolist())
print(unique_new_weight_column)
print(len(unique_new_weight_column))
```

```python
df2['weight'] = new_weight_column
df2.head()
```
